### Handling `Null` Values

```python
df.isnull().sum()
```

```python
df["number_of_cells"] = df["number_of_cells"].fillna(int(df["number_of_cells"].mean()))
df["towing_capacity_kg"] = df["towing_capacity_kg"].fillna(int(df["towing_capacity_kg"].mean()))
df["torque_nm"] = df["torque_nm"].fillna(int(df["torque_nm"].mean()))
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

### Deleteing least important Features 

```python
df.drop(columns=["source_url","model","battery_type","fast_charge_port"],inplace=True)
```

# ML Model Preparation  

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error , mean_absolute_error ,r2_score

```

### Encoding Catogircal Columns  

```python
for column in df.select_dtypes(["object"]):
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    

df.head()
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(18,12))
sns.heatmap(corr,annot=True,fmt="0.01f")
plt.show()
```

### Scaling Data with StandardScaler 

```python
for column in df.columns:
    if column != "acceleration_0_100_s":
        le = StandardScaler()
        df[column] = le.fit_transform(df[[column]])
```

### Split Data to **Train** and **Test**

```python
x = df.drop(['acceleration_0_100_s'], axis=1)
y = df['acceleration_0_100_s']
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=42,shuffle=True)
```

### Support Vector Machine 

```python
model = SVR(C=0.3)
model.fit(x_train,y_train)
y_predicted = model.predict(x_test)
print(f"Model Accurecy = {model.score(x_test,y_test)* 100:0.2f}")
print(f"Model MSE = {mean_squared_error(y_predicted,y_test):0.3f}")
print(f"Model MAE = {mean_absolute_error(y_predicted,y_test):0.3f}")
print(f"Model r2 Score = {r2_score(y_predicted,y_test):0.3f}")

```

### Linear Regression 

```python
model = LinearRegression()
model.fit(x_train,y_train)
y_predicted = model.predict(x_test)
print(f"Model Accurecy = {model.score(x_test,y_test)* 100:0.2f}")
print(f"Model MSE = {mean_squared_error(y_predicted,y_test):0.3f}")
print(f"Model MAE = {mean_absolute_error(y_predicted,y_test):0.3f}")
print(f"Model r2 Score = {r2_score(y_predicted,y_test):0.3f}")
```

### Decision Tree

```python
model = DecisionTreeRegressor(max_depth=2)
model.fit(x_train,y_train)
y_predicted = model.predict(x_test)
print(f"Model Accurecy = {model.score(x_test,y_test)* 100:0.2f}")
print(f"Model MSE = {mean_squared_error(y_predicted,y_test):0.3f}")
print(f"Model MAE = {mean_absolute_error(y_predicted,y_test):0.3f}")
print(f"Model r2 Score = {r2_score(y_predicted,y_test):0.3f}")
```
